```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains

import pandas as pd
import xlwings as xw
import time
```

```python
chrome_options = Options()
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument('disable-notifications')

# Medica
# driver = webdriver.Chrome(ChromeDriverManager().install())
chrome_driver_path = "chromedriver.exe"
driver = webdriver.Chrome(executable_path=chrome_driver_path, options = chrome_options)
url = 'https://events.jspargo.com/adlm24/Public/Exhibitors.aspx?Index=All'
driver.get(url)
driver.maximize_window()
time.sleep(1)
accept_xpath = '//*[@id="adroll_allow_all"]'
accept = driver.find_element(By.XPATH, accept_xpath)
accept.click()
```

```python
# Locate the table and tbody element
tbody = driver.find_element(By.XPATH, '//*[@id="ctl00_dvContent"]/div[8]/div[1]/div/div/div[3]/table/tbody')
rows = tbody.find_elements(By.TAG_NAME, 'tr')
data = []

for row in rows:
    cols = row.find_elements(By.TAG_NAME, 'td')
    cols = [col.text.strip() for col in cols]
    data.append(cols)
    print(cols)

# If the table has column headers, extract them
headers = [header.text.strip() for header in table.find_elements(By.TAG_NAME, 'th')]
df = pd.DataFrame(data, columns=headers) if headers else pd.DataFrame(data)
```

```python
namen_liste = []
booth_liste = []

for row in data:
    namen_liste.append(row[1])
    booth_liste.append(row[2])
```

```python
# Excel ausfüllen
wb = xw.Book('Ausstellerliste_Vorlage.xlsx')
sheet = wb.sheets[0]

start_row = 3

A_col_range = sheet.range('A{}:A{}'.format(start_row, start_row+len(namen_liste)))
A_col_range.value = [[value] for value in namen_liste]

I_col_range = sheet.range('I{}:I{}'.format(start_row, start_row+len(booth_liste)))
I_col_range.value = [[value] for value in booth_liste]
```

```python
urls = []

for row in rows:
    cols = row.find_elements(By.TAG_NAME, 'td')
    if len(cols) > 1:  # Check if there are at least two <td> elements
        try:
            second_td = cols[1]  # The second <td> element (index 1)
            link = second_td.find_element(By.TAG_NAME, 'a')
            href = link.get_attribute('href')
            if href:
                urls.append(href)
                print(href)
        except Exception as e:
            # Handle the case where no <a> tag is found or other issues
            print(f"Error processing row: {e}")
            continue
```

```python
# add hyperlinks namen
for i in range(len(urls)):
    row = start_row+i    
    cell_row, cell_column = row, 1
    cell_address = (cell_row, cell_column)
    sheet.range(cell_address).add_hyperlink(urls[i], text_to_display=namen_liste[i])
    sheet.range(cell_address).api.Font.Bold = True
    sheet.range(cell_address).font.color = (0, 0, 0)
    sheet.range(cell_address).api.Font.Size = 14
    sheet.range(cell_address).api.Font.Name = 'Verdana'
```

```python
# add hyperlinks booth
urls_booth = []
for row in rows:
    cols = row.find_elements(By.TAG_NAME, 'td')
    if len(cols) > 1:  # Check if there are at least two <td> elements
        try:
            second_td = cols[2]  # The second <td> element (index 1)
            link = second_td.find_element(By.TAG_NAME, 'a')
            href = link.get_attribute('href')
            if href:
                urls_booth.append(href)
                print(href)
        except Exception as e:
            # Handle the case where no <a> tag is found or other issues
            print(f"Error processing row: {e}")
            continue
```

```python
for i in range(len(urls_booth)):
    row = start_row+i    
    cell_row, cell_column = row, 9
    cell_address = (cell_row, cell_column)
    sheet.range(cell_address).add_hyperlink(urls_booth[i], text_to_display=booth_liste[i])
    sheet.range(cell_address).api.Font.Bold = True
    sheet.range(cell_address).font.color = (0, 0, 0)
    sheet.range(cell_address).api.Font.Size = 11
    sheet.range(cell_address).api.Font.Name = 'Verdana'
```
